```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
from fredapi import Fred
```

```python
#Data Pull From FRED API

rates = ['DGS1MO', 'DGS3MO', 'DGS6MO', 'DGS1', 'DGS2', 'DGS3', 'DGS5',
         'DGS7', 'DGS10', 'DGS20', 'DGS30']#Treasury
```

```python
#Set api_key
fred = Fred(api_key='set_api_key')
```

```python
#Initiate pd.DataFrame
df_rates = pd.DataFrame(fred.get_series(rates[0]))
df_rates['Symbol'] = np.repeat(rates[0], df_rates.shape[0])
df_rates.columns = ['IR', 'Symbol']
```

```python
#Loop through remaining Symbols (Treasury Rates)
for i in rates[1:len(rates)]:
    
    add = pd.DataFrame(fred.get_series(i))
    add['Symbol'] = np.repeat(i, add.shape[0])
    add.columns = ['IR', 'Symbol']
    df_rates = df_rates.append(add)
    
df_rates = df_rates.reset_index()
df_rates = df_rates.rename(columns = {"index": "Date"})
```

```python
#Graph (Subplot)
fig, axes = plt.subplots(nrows = 4, ncols = 3,figsize = (15,10))

axes = axes.ravel()
for x in range(len(rates)):
        
    axes[x].plot(df_rates.loc[df_rates['Symbol'] == rates[x]]['Date'], 
    df_rates.loc[df_rates['Symbol'] == rates[x]]['IR'])   
    axes[x].set_xlabel('Time')
    axes[x].set_ylabel('Treasury Yield %')
    axes[x].set_title(rates[x])
    
    plt.tight_layout()  

#Cut off last empty plot as odd number of rates (11)    
fig.delaxes(ax= axes[11])
```
